```python
rmap_name_id = {}
rmap_id_tid = {}
with open('map/data/regions-table.txt') as table_file:
    lines = table_file.readlines()
    for l in lines:
        v = l.split('\t')
        name = v[2]
        tid = int(v[1]) # id конституции
        nid = int(v[0]) # порядок
        
        rmap_name_id[name] = tid
        rmap_id_tid[nid] = tid
rmap_name_id['Республика Татарстан'] = rmap_name_id['Республика Татарстан (Татарстан)']
```

```python
import difflib
short_code_to_id_law = {}
for r in regions.iterrows():
    try:
        if r[1][2] in short_code_to_id_law:
            print(r[1][2])
        short_code_to_id_law[r[1][2]] = rmap_name_id[difflib.get_close_matches(r[1][0], rmap_name_id.keys())[0]]
    except:
        print(r)
```

Байконур не является субъектом РФ

```python
short_code_to_id_law
```

```python
ans = {}
for ((region, week), c) in cnt.iteritems():
    id_law = 0
    if region in short_code_to_id_law:
        id_law = short_code_to_id_law[region]
    else: continue
    if region not in ans:
        ans[id_law] = {'val':{}, 'proc':{}}
    ans[id_law]['val'][week] = c
```

```python
for r in ans.keys():
    for w,c in ans[r]['val'].items():
        mean = 0;
        keys = list(ans[r]['val'].keys())
        for i in range(len(keys)):
            mean += ans[r]['val'][keys[i]]
        mean /= len(keys)
        
        ans[r]['proc'][w] = ans[r]['val'][w] / mean
```
