```python
rmap_name_id = {}
rmap_id_tid = {}
with open('regions-table.txt') as table_file:
    lines = table_file.readlines()
    for l in lines:
        v = l.split('\t')
        name = v[2]
        tid = int(v[1]) # id конституции
        nid = int(v[0]) # порядок
        
        rmap_name_id[name] = tid
        rmap_id_tid[nid] = tid
```

```python
import difflib

ans = {}
dates = set()

for l in frame.iterrows():
    id_law = rmap_name_id[ difflib.get_close_matches(l[1][1], rmap_name_id.keys())[0] ]
    dates.add(l[1][0])
    if id_law not in ans:
        ans[id_law] = {'arr':{}}
        
    ans[id_law]['arr'][date_to_week[l[1][0]]] = {'c':l[1][2],'ac':l[1][3],'nc':l[1][8],'ah':l[1][4],'nh':l[1][5],'ad':l[1][6],'nd':l[1][7]}
    ans[id_law]['name']=l[1][1]
    ans[id_law]['count']=l[1][9]
```
